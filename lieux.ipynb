### Chargement des librarires :

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

### Chargement des bases de données liées aux lieux pour les années 2017 à 2021 :


```python
df_lieux_2021 = pd.read_csv("data/lieux-2021.csv", sep=';')
df_lieux_2020 = pd.read_csv("data/lieux-2020.csv", sep=';')
df_lieux_2019 = pd.read_csv("data/lieux-2019.csv", sep=';')
df_lieux_2018 = pd.read_csv("data/lieux-2018.csv", sep=',', low_memory=False)
df_lieux_2017 = pd.read_csv("data/lieux-2017.csv", sep=',', low_memory=False)
```

### Fusion et nettoyage des dataframes :

```python
df = [df_lieux_2017, df_lieux_2018, df_lieux_2019, df_lieux_2020, df_lieux_2021]
lieux_2017_2021 = pd.concat(df)
```

```python
sns.heatmap(lieux_2017_2021.isna())
```

```python
#remplacement des -1 (non renseigné) par des NaN(Not a Number)
lieux_2017_2021 = lieux_2017_2021.replace(-1, np.nan)
```

```python
sns.heatmap(lieux_2017_2021.isna());
```

```python
lieux_2017_2021 = lieux_2017_2021.replace(0, np.nan)
```

```python
sns.heatmap(lieux_2017_2021.isna());
```

```python
lieux_2017_2021.to_csv('merged_tables/lieux_2017_2021.csv', index=False)
```
