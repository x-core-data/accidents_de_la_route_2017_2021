### Chargement des librarires nécessaires à l'exploitation des données :

```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
```

### Chargement des bases de données liées aux véhicules avec la méthode "read_csv" de pandas pour les années 2017 à 2021 :


```python
df_vehicules_2021 = pd.read_csv("data/vehicules-2021.csv", sep=';')
df_vehicules_2020 = pd.read_csv("data/vehicules-2020.csv", sep=';')
df_vehicules_2019 = pd.read_csv("data/vehicules-2019.csv", sep=';')
df_vehicules_2018 = pd.read_csv("data/vehicules-2018.csv", sep=',', low_memory=False)
df_vehicules_2017 = pd.read_csv("data/vehicules-2017.csv", sep=',', low_memory=False)
```

```python
df_vehicules_2017
```

### Ajouter une colonne 'année_accident' pour chaque dataframe :

```python
df_vehicules_2021 = df_vehicules_2021.assign(annee_accident=2021)
df_vehicules_2020 = df_vehicules_2020.assign(annee_accident=2020)
df_vehicules_2019 = df_vehicules_2019.assign(annee_accident=2019)
df_vehicules_2018 = df_vehicules_2018.assign(annee_accident=2018)
df_vehicules_2017 = df_vehicules_2017.assign(annee_accident=2017)
```

### Utiliser la méthode concat afin de rassembler les cinq dataframes :

```python
df = [df_vehicules_2017, df_vehicules_2018, df_vehicules_2019, df_vehicules_2020, df_vehicules_2021]
vehicules_2017_2021 = pd.concat(df)
```

```python
vehicules_2017_2021
```

```python
#remplacement des -1 (non renseigné) par des NaN(Not a Number)
vehicules_2017_2021 = vehicules_2017_2021.replace(-1, np.nan)
```

```python
vehicules_2017_2021.to_csv('vehicules_2017_2021.csv', index=False)
```
